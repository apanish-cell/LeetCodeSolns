# Two Sum

```python
class Solution:
  # twoSum Problem
  def twoSum(self, nums: list[int], target: int) -> list[int]:
    self.nums = nums
    self.target = target

    for i in range(len(self.nums)):
      # Start j at i+1 to avoid using the same element twice
      for j in range(i+1, len(self.nums)):
        x = self.nums[i]
        if self.nums[j] == self.target - x:
          return [i,j]

```

```python
nums = [3,3]
target = int(input("Enter a Number: "))

solution_instance = Solution()
result = solution_instance.twoSum(nums, target)
print(result)
```

# Palindrone

```python
class Solution1:
    # isPalindrome Problem
    def isPalindrome(self, x: int) -> bool:
      self.x = x
      x_str = str(self.x)
      rev_string = x_str[::-1]

      if x_str == rev_string:
        return True
      else:
        return False
```

```python
x = 130
obj = Solution1()
pali= obj.isPalindrome(x)
print(pali)
```

# Roman to Integers

```python
class Solution2:
    def romanToInt(self, s: str) -> int:
        self.s = s

        symbols = {
            "I": 1,
            "V": 5,
            "X": 10,
            "L": 50,
            "C": 100,
            "D": 500,
            "M": 1000
        }

        return symbols.get(self.s)
```

```python
i = "V"
obj1 = Solution2()
rom2int = obj1.romanToInt(i)
print(rom2int)
```
